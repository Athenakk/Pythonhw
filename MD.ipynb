## 利用網站找出所要商品的資訊

```python
import requests
import json

res=requests.get('https://ecshweb.pchome.com.tw/search/v3.3/all/results?q=airpods&page=1&sort=sale/dc')
```

```python
data=json.loads(res.text) #把資訊轉為json形式
```

```python
data
```

```python
wd=data['prods']
```

```python
len(wd)
```

```python
for product in wd:
    print(product['name'])
    print(product['price'])
```
